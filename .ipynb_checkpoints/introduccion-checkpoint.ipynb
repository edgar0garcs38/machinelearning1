### 3.0.2 ¿Qué es un intérprete?<br>
• Un programa que ejecuta otros programas.<br>
• Una capa de software entre tu código y el hardware que lo ejecuta.<br>
• Debe estar instalado en tu ordenador para poder ejecutar código Python.<br>
• Para la especificación de Python, existen varias implementaciones:<br>
– CPython (implementación en C). Es el más común.<br>
– Jython (implementación en Java).<br>
– IronPython (implementación en .NET).<br>

### 3.0.8 REPL<br>
Sistema interactivo para comunicarse con el ordenador en un leguaje, Python. Se debe cumplir: *
Read. El ordenador pueda leer unidades como entrada * Evaluate. El código pueda ser procesado
* Print. Los resultados puedan verse * Loop. Continuar con la conversación
### 3.0.9 Ficheros<br>
• Permite almacenar programas.<br>
• Ficheros de texto con instrucciones Python.<br>
– No olvidar shebang en Linux -> #!/usr/bin/env python3<br>
• Terminología (varía segun fuentes):<br>
– Scripts o programas: programa principal.<br>
– Módulos: ficheros importados desde otros ficheros.<br>
• Se puede lanzar pasando nombre de fichero a comando python.<br>
> python ./script1.py<br>
Hello world<br>
• Otra alternativa (a partir the Python 3.3) es:<br>
> py ./script1.py<br>
Hello world<br>
• O incluso:<br>
> ./script1.py<br>
Hello world<br>
• También es posible hacer doble-click sobre fichero .py.<br>

### 3.0.12 Performance Sample, NumPy vs Plain Python

```python
import numpy as np
my_arr = np.arange(10000000) # Utilizando NumPy Arrays
my_list = list(range(10000000)) # Utilizando Listas
print(my_list)
```

```python
%time for _ in range(10): my_arr2 = my_arr * 2
%time for _ in range(10): my_list2 = [x * 2 for x in my_list]
```
